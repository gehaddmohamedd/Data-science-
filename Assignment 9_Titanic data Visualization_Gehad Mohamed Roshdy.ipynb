```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pywaffle import Waffle
from wordcloud import WordCloud
```

```python
df_total=pd.read_csv(r"C:\Users\Masria For PCs\Downloads\Titanic.csv")
df_total.head()
```

```python
df_total.info
```

```python
sns.countplot(x=df_total['Pclass'])
plt.title('Numberof Passengers by Class') 
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
data = df_total['Sex'].value_counts().to_dict()
fig=plt.figure(
    FigureClass=Waffle,
    rows=10,
    columns=20,
    values=data,
    title={"label": " Gender Waffle chart", "loc":"center"},
    legend={"loc": "lower right", "bbox_to_anchor": (1.3, 0)}
)
plt.show
```

```python
text = df_total['Name'].astype(str).str.cat(sep=" ")
wordcloud= WordCloud(width=800, height=400, background_color="white").generate(text)
plt.figure(figsize=(10,5))
plt.title('Names of passengers') 
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
sns.histplot(df_total['Age'], bins=30)
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(df_total['Fare'], bins=30)
plt.title('Fare Distribution of Titanic Passengers')
plt.xlabel('Fair')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x=df_total['Fare'])
plt.title('Boxplot of Passenger Fares')
plt.xlabel('Fare')
plt.show()
```

```python
df_total['Survived'].value_counts().plot(kind='pie', autopct='%1.1f%%', labels=['Did Not Survive', 'Survived'])
plt.title('Survivals')
plt.ylabel('')
plt.show()
```

```python
sns.scatterplot(x=df_total['Age'], y=df_total['Fare'], alpha=0.5)
plt.title('Scatter Plot of Fare vs. Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
sns.scatterplot(x=df_total['Fare'], y=df_total['Pclass'], alpha=0.5)
plt.title('Scatter Plot of Fare vs. Pclass')
plt.xlabel('Fare')
plt.ylabel('Pclass')
plt.show()
```
